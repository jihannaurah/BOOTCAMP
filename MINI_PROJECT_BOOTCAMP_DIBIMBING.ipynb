```python
import pandas as pd
df = pd.read_csv('data.csv', encoding='latin1')
df.head()
```

DATA CLEANING

```python
df = df.drop_duplicates()
```

```python
df = df.dropna(subset=['CustomerID'])
```

```python
print(df.shape)
```

```python
df.isnull().sum()
```

```python
df = df.reset_index(drop=True)
```

```python
df.info()
```

FILTERING "QUANTITY, UNITPRICE"

```python
df = df[df['Quantity'] > 0]
```

```python
df = df[df['UnitPrice'] > 0]
```

```python
df = df.reset_index(drop=True)
```

```python
print("Hasil:")
print(df.shape)
```

```python
print("\ninfo detail:")
df.info()
```

MASKING

```python
df['CustomerID'] = df['CustomerID'].astype(str).str[:2] + '***' + df['CustomerID'].astype(str).str[-2:]
df['StockCode'] = df['StockCode'].astype(str).str[:2] + '***' + df['StockCode'].astype(str).str[-2:]
df['InvoiceNo'] = df['InvoiceNo'].astype(str).str[:2] + '***' + df['InvoiceNo'].astype(str).str[-2:]
print("Data Masking")
df.sample(20)
```

**DATA DICTIONARY**
| Nama Kolom | Tipe | Penjelasan Singkat | Contohnya |
| :--- | :--- | :--- | :--- |
| **InvoiceNo** | Teks | Kode transaksi. Kalau ada huruf 'C' di depan, berarti itu barang yang dicancel/retur. | 536365 |
| **StockCode** | Teks | Kode buat tiap barang. (masking). | 85*** |
| **Description** | Teks | Nama barang atau penjelasan produk yang dibeli pelanggan. | BOX OF 6 MINI VINTAGE CRACKERS... |
| **Quantity** | Angka | Jumlah barang yang dibeli. Udah di filter jadi cuma angka yang di atas 0. | 6 |
| **InvoiceDate** | Teks | Tanggal sama jam kapan transaksi itu terjadi. | 12/1/2010 8:26 |
| **UnitPrice** | Desimal | Harga satu barang dalam Poundsterling (£). | 2.55 |
| **CustomerID** | Teks | ID Pelanggan. Udah di sensor (masking) juga buat jaga data sensitif. | 17*** |
| **Country** | Teks | Negara tempat belanja atau asal pelanggannya. | United Kingdom |
